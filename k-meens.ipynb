```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score, silhouette_score, silhouette_samples

# ConfiguraciÃ³n para grÃ¡ficos en espaÃ±ol
plt.rcParams['figure.figsize'] = (10, 6)

# Establecer semilla para reproducibilidad
np.random.seed(42)

print("=== TUTORIAL INTERACTIVO: K-MEANS CON ANÃLISIS DE MÃ‰TRICAS ===\n")

# 1. GENERACIÃ“N DEL DATASET
print("ğŸ¯ PASO 1: Generando dataset sintÃ©tico...")
n_samples = 300
n_centers = 4
X, y_true = make_blobs(n_samples=n_samples, 
                       centers=n_centers, 
                       cluster_std=1.0,
                       center_box=(-10, 10),
                       random_state=42)

print(f"ğŸ“Š Forma del dataset: {X.shape}")
print(f"ğŸ¯ NÃºmero real de clusters: {n_centers}")
print("ğŸ’¡ make_blobs() crea puntos agrupados naturalmente - perfecto para validar nuestro algoritmo\n")

# 2. APLICACIÃ“N DE K-MEANS
print("ğŸ” PASO 2: Aplicando K-means clustering...")
k = 4  # NÃºmero de clusters que queremos encontrar
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
y_pred = kmeans.fit_predict(X)

cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_
inertia = kmeans.inertia_

print(f"ğŸ“ Centroides encontrados:\n{cluster_centers.round(2)}")
print(f"ğŸ“ Inercia: {inertia:.2f}")
print("ğŸ’¡ INERCIA: Suma de distancias al cuadrado de cada punto a su centroide")
print("   - Valores mÃ¡s BAJOS = clusters mÃ¡s compactos")
print("   - Es la funciÃ³n objetivo que K-means minimiza\n")

# 3. MÃ‰TRICAS DE EVALUACIÃ“N
print("ğŸ“Š PASO 3: Evaluando la calidad del clustering...")

# Adjusted Rand Index
ari_score = adjusted_rand_score(y_true, y_pred)
print(f"ğŸ¯ Adjusted Rand Index (ARI): {ari_score:.3f}")
print("ğŸ’¡ ADJUSTED RAND INDEX:")
print("   - Compara clusters predichos vs clusters reales")
print("   - Rango: -1 a 1 (1 = perfecto, 0 = aleatorio, <0 = peor que aleatorio)")
print(f"   - Nuestro resultado {ari_score:.3f} indica: ", end="")
if ari_score > 0.8:
    print("Â¡EXCELENTE concordancia!")
elif ari_score > 0.5:
    print("Buena concordancia")
else:
    print("Concordancia regular")

# Silhouette Score
silhouette_avg = silhouette_score(X, y_pred)
print(f"\nğŸ” PuntuaciÃ³n Promedio de Silueta: {silhouette_avg:.3f}")
print("ğŸ’¡ PUNTUACIÃ“N DE SILUETA:")
print("   - Mide quÃ© tan similar es un punto a su cluster vs otros clusters")
print("   - Rango: -1 a 1 (1 = muy bien agrupado, 0 = en frontera, -1 = mal agrupado)")
print(f"   - Nuestro resultado {silhouette_avg:.3f} indica: ", end="")
if silhouette_avg > 0.7:
    print("Â¡EXCELENTE separaciÃ³n!")
elif silhouette_avg > 0.5:
    print("Buena separaciÃ³n")
elif silhouette_avg > 0.3:
    print("SeparaciÃ³n regular")
else:
    print("SeparaciÃ³n pobre")

print("\n" + "="*60)

# 4. VISUALIZACIÃ“N PRINCIPAL
print("ğŸ“ˆ PASO 4: Visualizando resultados...")
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# GrÃ¡fico de clusters verdaderos
scatter1 = ax1.scatter(X[:, 0], X[:, 1], c=y_true, cmap='viridis', alpha=0.7)
ax1.set_title('Clusters Verdaderos', fontsize=14, fontweight='bold')
ax1.set_xlabel('CaracterÃ­stica 1')
ax1.set_ylabel('CaracterÃ­stica 2')
ax1.grid(True, alpha=0.3)
plt.colorbar(scatter1, ax=ax1)

# GrÃ¡fico de resultados K-means
scatter2 = ax2.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', alpha=0.7)
ax2.scatter(cluster_centers[:, 0], cluster_centers[:, 1], 
           c='red', marker='x', s=300, linewidths=4, label='Centroides')
ax2.set_title('Resultados K-means', fontsize=14, fontweight='bold')
ax2.set_xlabel('CaracterÃ­stica 1')
ax2.set_ylabel('CaracterÃ­stica 2')
ax2.grid(True, alpha=0.3)
ax2.legend()
plt.colorbar(scatter2, ax=ax2)

plt.tight_layout()
plt.show()

# 5. MÃ‰TODO DEL CODO (ELBOW METHOD)
print("ğŸ“Š PASO 5: Aplicando el MÃ©todo del Codo para encontrar k Ã³ptimo...")
k_range = range(1, 11)
inertias = []
silhouette_scores = []

for k in k_range:
    if k == 1:
        # Silhouette no se puede calcular con 1 cluster
        inertias.append(KMeans(n_clusters=k, random_state=42, n_init=10).fit(X).inertia_)
        silhouette_scores.append(0)
    else:
        kmeans_temp = KMeans(n_clusters=k, random_state=42, n_init=10)
        y_pred_temp = kmeans_temp.fit_predict(X)
        inertias.append(kmeans_temp.inertia_)
        silhouette_scores.append(silhouette_score(X, y_pred_temp))

# GrÃ¡ficos del mÃ©todo del codo y silueta
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# MÃ©todo del Codo
ax1.plot(k_range, inertias, 'bo-', linewidth=2, markersize=8)
ax1.set_xlabel('NÃºmero de Clusters (k)', fontsize=12)
ax1.set_ylabel('Inercia', fontsize=12)
ax1.set_title('MÃ©todo del Codo', fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.axvline(x=4, color='red', linestyle='--', alpha=0.7, label='k=4 (Ã³ptimo)')
ax1.legend()

# MÃ©todo de la Silueta
ax2.plot(k_range[1:], silhouette_scores[1:], 'ro-', linewidth=2, markersize=8)
ax2.set_xlabel('NÃºmero de Clusters (k)', fontsize=12)
ax2.set_ylabel('PuntuaciÃ³n Promedio de Silueta', fontsize=12)
ax2.set_title('MÃ©todo de la Silueta', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3)
# Encontrar el k con mejor silueta
best_k = k_range[1:][np.argmax(silhouette_scores[1:])]
ax2.axvline(x=best_k, color='red', linestyle='--', alpha=0.7, label=f'k={best_k} (mÃ¡xima silueta)')
ax2.legend()

plt.tight_layout()
plt.show()

print("ğŸ’¡ INTERPRETACIÃ“N DE LOS MÃ‰TODOS:")
print("ğŸ”¹ MÃ‰TODO DEL CODO:")
print("   - Busca el punto donde la inercia deja de decrecer significativamente")
print("   - El 'codo' en la curva indica el k Ã³ptimo")
print(f"   - En este caso, k=4 muestra el codo mÃ¡s claro")

print("\nğŸ”¹ MÃ‰TODO DE LA SILUETA:")
print("   - Busca el k que maximiza la puntuaciÃ³n promedio de silueta")
print(f"   - El k Ã³ptimo segÃºn silueta es: k={best_k}")
print(f"   - PuntuaciÃ³n mÃ¡xima: {max(silhouette_scores[1:]):.3f}")

# 6. ANÃLISIS DETALLADO DE SILUETA - EXPLICACIÃ“N PARA DUMMIES
print(f"\nğŸ“Š PASO 6: Â¿QuÃ© es la Silueta? - ExplicaciÃ³n sÃºper simple")

print("ğŸ¤” IMAGINE QUE USTED ES UN PUNTO EN EL ESPACIO:")
print("   1. Mire a sus 'vecinos' en SU cluster â†’ calcule distancia promedio = 'a'")
print("   2. Mire al cluster MÃS CERCANO â†’ calcule distancia promedio = 'b'")
print("   3. Su valor de silueta = (b - a) / max(a, b)")
print("")
print("ğŸ¯ Â¿QUÃ‰ SIGNIFICA ESTO?")
print("   â€¢ Si b >> a â†’ Silueta cerca de +1 â†’ Â¡ESTÃ MUY BIEN UBICADO!")
print("   â€¢ Si b â‰ˆ a  â†’ Silueta cerca de 0  â†’ EstÃ¡ en la 'frontera'")
print("   â€¢ Si b << a â†’ Silueta cerca de -1 â†’ Â¡ESTÃ MAL UBICADO!")
print("")

# DemostraciÃ³n con un punto especÃ­fico
punto_ejemplo = X[0]
cluster_ejemplo = y_pred[0]
print(f"ğŸ” EJEMPLO PRÃCTICO con el punto {punto_ejemplo.round(2)}:")
print(f"   - EstÃ¡ asignado al cluster {cluster_ejemplo}")

# Calcular 'a' - distancia promedio dentro de su cluster
puntos_mismo_cluster = X[y_pred == cluster_ejemplo]
distancias_mismo_cluster = np.sqrt(np.sum((puntos_mismo_cluster - punto_ejemplo)**2, axis=1))
a = np.mean(distancias_mismo_cluster)

# Calcular 'b' - distancia promedio al cluster mÃ¡s cercano
distancias_otros_clusters = []
for cluster_id in range(k):
    if cluster_id != cluster_ejemplo:
        puntos_otro_cluster = X[y_pred == cluster_id]
        dist_promedio = np.mean(np.sqrt(np.sum((puntos_otro_cluster - punto_ejemplo)**2, axis=1)))
        distancias_otros_clusters.append(dist_promedio)

b = min(distancias_otros_clusters)
silueta_ejemplo = (b - a) / max(a, b)

print(f"   - Distancia promedio a SU cluster (a): {a:.2f}")
print(f"   - Distancia promedio al cluster MÃS CERCANO (b): {b:.2f}")
print(f"   - Su valor de silueta: ({b:.2f} - {a:.2f}) / {max(a,b):.2f} = {silueta_ejemplo:.3f}")

if silueta_ejemplo > 0.5:
    print("   - âœ… Â¡Este punto estÃ¡ MUY bien ubicado!")
elif silueta_ejemplo > 0:
    print("   - âœ… Este punto estÃ¡ bien ubicado")
else:
    print("   - âŒ Este punto podrÃ­a estar mejor en otro cluster")

print("\n" + "="*70)
print("ğŸ“Š AHORA VEAMOS TODOS LOS PUNTOS...")

sample_silhouette_values = silhouette_samples(X, y_pred)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))

# GrÃ¡fico de silueta CON EXPLICACIONES DETALLADAS
y_lower = 10
colors = plt.cm.viridis(np.linspace(0, 1, k))

print(f"ğŸ“ˆ LEYENDO EL GRÃFICO DE SILUETA:")
for i in range(k):
    cluster_silhouette_values = sample_silhouette_values[y_pred == i]
    
    size_cluster_i = cluster_silhouette_values.shape[0]
    
    # Verificar que el cluster no estÃ© vacÃ­o
    if size_cluster_i == 0:
        print(f"ğŸ”¹ Cluster {i}: Â¡VACÃO! (0 puntos asignados)")
        print(f"   - âš ï¸  Este cluster no tiene puntos asignados")
        continue
        
    cluster_silhouette_values.sort()
    y_upper = y_lower + size_cluster_i
    
    # EstadÃ­sticas del cluster (solo si no estÃ¡ vacÃ­o)
    min_sil = cluster_silhouette_values.min()
    max_sil = cluster_silhouette_values.max()
    avg_sil = cluster_silhouette_values.mean()
    
    print(f"ğŸ”¹ Cluster {i}: {size_cluster_i} puntos")
    print(f"   - Silueta mÃ­n/mÃ¡x/promedio: {min_sil:.3f} / {max_sil:.3f} / {avg_sil:.3f}")
    
    # Evaluar calidad del cluster
    puntos_negativos = np.sum(cluster_silhouette_values < 0)
    if puntos_negativos > 0:
        print(f"   - âš ï¸  {puntos_negativos} puntos con silueta negativa (mal ubicados)")
    else:
        print(f"   - âœ… Todos los puntos bien ubicados")
    
    if avg_sil > silhouette_avg:
        print(f"   - ğŸ‘ Cluster MEJOR que el promedio")
    else:
        print(f"   - ğŸ‘ Cluster PEOR que el promedio")
    
    ax1.fill_betweenx(np.arange(y_lower, y_upper),
                      0, cluster_silhouette_values,
                      facecolor=colors[i], edgecolor=colors[i], alpha=0.7)
    
    # Etiqueta del cluster con estadÃ­sticas
    ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, 
             f'C{i}\n({size_cluster_i}p)', fontsize=10, fontweight='bold')
    y_lower = y_upper + 10

ax1.set_xlabel('Valores de Silueta\n(MÃ¡s a la derecha = Mejor ubicado)', fontsize=12)
ax1.set_ylabel('Puntos agrupados por Cluster', fontsize=12)
ax1.set_title('ğŸ¯ GrÃ¡fico de Silueta - Â¿QuÃ© tan bien estÃ¡n ubicados los puntos?', 
              fontsize=13, fontweight='bold')

# LÃ­nea vertical para el promedio CON EXPLICACIÃ“N
ax1.axvline(x=silhouette_avg, color="red", linestyle="--", linewidth=3,
            label=f'Promedio = {silhouette_avg:.3f}')
ax1.axvline(x=0, color="orange", linestyle=":", linewidth=2,
            label='LÃ­mite Bueno/Malo')

# Agregar zonas de interpretaciÃ³n
ax1.axvspan(-1, 0, alpha=0.1, color='red', label='Zona MALA')
ax1.axvspan(0, 0.3, alpha=0.1, color='yellow', label='Zona REGULAR')
ax1.axvspan(0.3, 1, alpha=0.1, color='green', label='Zona BUENA')

ax1.legend(loc='lower right')
ax1.grid(True, alpha=0.3)
ax1.set_xlim(-0.3, 1)

# GrÃ¡fico de clusters con colores de silueta Y EXPLICACIONES
scatter = ax2.scatter(X[:, 0], X[:, 1], c=sample_silhouette_values, 
                     cmap='RdYlGn', alpha=0.8, s=60)  # RdYlGn: rojo=malo, amarillo=regular, verde=bueno
ax2.scatter(cluster_centers[:, 0], cluster_centers[:, 1], 
           c='blue', marker='x', s=400, linewidths=5, label='Centroides')

# Marcar puntos problemÃ¡ticos
puntos_malos = sample_silhouette_values < 0
if np.any(puntos_malos):
    ax2.scatter(X[puntos_malos, 0], X[puntos_malos, 1], 
               s=100, facecolors='none', edgecolors='red', linewidths=2,
               label=f'{np.sum(puntos_malos)} puntos mal ubicados')

ax2.set_title('ğŸ¨ Mapa de Calidad: Verde=Bien, Amarillo=Regular, Rojo=Mal', 
              fontsize=13, fontweight='bold')
ax2.set_xlabel('CaracterÃ­stica 1')
ax2.set_ylabel('CaracterÃ­stica 2')
ax2.legend()

# Barra de colores con interpretaciÃ³n
cbar = plt.colorbar(scatter, ax=ax2, label='Valor de Silueta')
cbar.ax.set_ylabel('ğŸ”´ Mal ubicado â† â†’ Bien ubicado ğŸŸ¢', rotation=270, labelpad=20)

plt.tight_layout()
plt.show()

print("\nğŸ’¡ RESUMEN PARA DUMMIES - Â¿CÃ“MO LEER ESTOS GRÃFICOS?")
print("="*70)
print("ğŸ“Š GRÃFICO DE LA IZQUIERDA (GrÃ¡fico de Silueta):")
print("   ğŸ”¹ Cada 'montaÃ±ita' horizontal = un cluster")
print("   ğŸ”¹ Altura de la montaÃ±ita = nÃºmero de puntos en ese cluster")
print("   ğŸ”¹ Ancho hacia la derecha = quÃ© tan bien ubicados estÃ¡n los puntos")
print("   ğŸ”¹ Si una montaÃ±ita cruza la lÃ­nea roja = cluster mejor que promedio")
print("   ğŸ”¹ Si hay partes a la IZQUIERDA del 0 = puntos mal ubicados")
print("")
print("ğŸ¨ GRÃFICO DE LA DERECHA (Mapa de Colores):")
print("   ğŸŸ¢ Verde = Puntos felices en su cluster")
print("   ğŸŸ¡ Amarillo = Puntos 'meh', podrÃ­an estar mejor")
print("   ğŸ”´ Rojo = Puntos tristes, estÃ¡n en el cluster equivocado")
print("   ğŸ’™ X Azules = Centros de cada cluster")
print("   ğŸ”´ CÃ­rculos rojos = Puntos definitivamente mal ubicados")

# Dar recomendaciones basadas en los resultados
puntos_negativos_total = np.sum(sample_silhouette_values < 0)
print(f"\nğŸ¯ DIAGNÃ“STICO DE NUESTRO CLUSTERING:")
if puntos_negativos_total == 0:
    print("âœ… Â¡PERFECTO! NingÃºn punto mal ubicado")
elif puntos_negativos_total < len(X) * 0.05:  # Menos del 5%
    print(f"âœ… MUY BUENO: Solo {puntos_negativos_total} puntos mal ubicados ({puntos_negativos_total/len(X)*100:.1f}%)")
elif puntos_negativos_total < len(X) * 0.1:   # Menos del 10%
    print(f"âš ï¸  BUENO: {puntos_negativos_total} puntos mal ubicados ({puntos_negativos_total/len(X)*100:.1f}%)")
else:
    print(f"âŒ REGULAR: Muchos puntos mal ubicados ({puntos_negativos_total/len(X)*100:.1f}%)")
    print("   ğŸ’¡ Sugerencia: Probar con diferente nÃºmero de clusters")

if silhouette_avg > 0.7:
    print("ğŸ† Calidad general: EXCELENTE")
elif silhouette_avg > 0.5:
    print("ğŸ‘ Calidad general: BUENA") 
elif silhouette_avg > 0.3:
    print("ğŸ¤· Calidad general: REGULAR")
else:
    print("ğŸ‘ Calidad general: POBRE - considere cambiar k")

print("="*70)

# 7. PREDICCIÃ“N EN NUEVOS PUNTOS
print(f"\nğŸ¯ PASO 7: Prediciendo clusters para nuevos puntos...")
new_points = np.array([[2, 2], [-8, 3], [6, -6]])
predicted_clusters = kmeans.predict(new_points)

print(f"ğŸ“ Nuevos puntos: {new_points}")
print(f"ğŸ¯ Clusters predichos: {predicted_clusters}")

# Calcular distancias a centroides para explicar la asignaciÃ³n
print("\nğŸ’¡ EXPLICACIÃ“N DE LAS PREDICCIONES:")
for i, point in enumerate(new_points):
    distances = np.sqrt(np.sum((cluster_centers - point)**2, axis=1))
    closest_centroid = np.argmin(distances)
    print(f"ğŸ“ Punto {i+1} {point}:")
    print(f"   - Asignado al cluster {predicted_clusters[i]}")
    print(f"   - Distancia al centroide mÃ¡s cercano: {distances[closest_centroid]:.2f}")

# VisualizaciÃ³n final con nuevos puntos
plt.figure(figsize=(10, 8))
scatter = plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', alpha=0.6, s=50)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], 
           c='red', marker='x', s=300, linewidths=4, label='Centroides')
plt.scatter(new_points[:, 0], new_points[:, 1], 
           c='black', marker='s', s=200, label='Nuevos Puntos', edgecolors='white', linewidths=2)

# Anotar los nuevos puntos
for i, point in enumerate(new_points):
    plt.annotate(f'P{i+1}\n(Cluster {predicted_clusters[i]})', 
                xy=point, xytext=(10, 10), textcoords='offset points',
                bbox=dict(boxstyle='round,pad=0.3', fc='yellow', alpha=0.7),
                fontsize=10, fontweight='bold')

plt.title('K-means: Clusters y Predicciones en Nuevos Puntos', fontsize=14, fontweight='bold')
plt.xlabel('CaracterÃ­stica 1')
plt.ylabel('CaracterÃ­stica 2')
plt.legend()
plt.colorbar(scatter, label='Cluster')
plt.grid(True, alpha=0.3)
plt.show()

print("\n" + "="*60)
print("ğŸ‰ RESUMEN DEL ANÃLISIS:")
print(f"âœ… NÃºmero Ã³ptimo de clusters: {k}")
print(f"âœ… Calidad del clustering (ARI): {ari_score:.3f}")
print(f"âœ… SeparaciÃ³n de clusters (Silueta): {silhouette_avg:.3f}")
print(f"âœ… Compacidad de clusters (Inercia): {inertia:.2f}")
print("âœ… El algoritmo funcionÃ³ correctamente identificando los 4 grupos naturales")
print("="*60)
```
