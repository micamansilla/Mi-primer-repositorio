```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score, silhouette_score, silhouette_samples

# Configuración para gráficos en español
plt.rcParams['figure.figsize'] = (10, 6)

# Establecer semilla para reproducibilidad
np.random.seed(42)

print("=== TUTORIAL INTERACTIVO: K-MEANS CON ANÁLISIS DE MÉTRICAS ===\n")

# 1. GENERACIÓN DEL DATASET
print("🎯 PASO 1: Generando dataset sintético...")
n_samples = 300
n_centers = 4
X, y_true = make_blobs(n_samples=n_samples, 
                       centers=n_centers, 
                       cluster_std=1.0,
                       center_box=(-10, 10),
                       random_state=42)

print(f"📊 Forma del dataset: {X.shape}")
print(f"🎯 Número real de clusters: {n_centers}")
print("💡 make_blobs() crea puntos agrupados naturalmente - perfecto para validar nuestro algoritmo\n")

# 2. APLICACIÓN DE K-MEANS
print("🔍 PASO 2: Aplicando K-means clustering...")
k = 4  # Número de clusters que queremos encontrar
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
y_pred = kmeans.fit_predict(X)

cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_
inertia = kmeans.inertia_

print(f"📍 Centroides encontrados:\n{cluster_centers.round(2)}")
print(f"📏 Inercia: {inertia:.2f}")
print("💡 INERCIA: Suma de distancias al cuadrado de cada punto a su centroide")
print("   - Valores más BAJOS = clusters más compactos")
print("   - Es la función objetivo que K-means minimiza\n")

# 3. MÉTRICAS DE EVALUACIÓN
print("📊 PASO 3: Evaluando la calidad del clustering...")

# Adjusted Rand Index
ari_score = adjusted_rand_score(y_true, y_pred)
print(f"🎯 Adjusted Rand Index (ARI): {ari_score:.3f}")
print("💡 ADJUSTED RAND INDEX:")
print("   - Compara clusters predichos vs clusters reales")
print("   - Rango: -1 a 1 (1 = perfecto, 0 = aleatorio, <0 = peor que aleatorio)")
print(f"   - Nuestro resultado {ari_score:.3f} indica: ", end="")
if ari_score > 0.8:
    print("¡EXCELENTE concordancia!")
elif ari_score > 0.5:
    print("Buena concordancia")
else:
    print("Concordancia regular")

# Silhouette Score
silhouette_avg = silhouette_score(X, y_pred)
print(f"\n🔍 Puntuación Promedio de Silueta: {silhouette_avg:.3f}")
print("💡 PUNTUACIÓN DE SILUETA:")
print("   - Mide qué tan similar es un punto a su cluster vs otros clusters")
print("   - Rango: -1 a 1 (1 = muy bien agrupado, 0 = en frontera, -1 = mal agrupado)")
print(f"   - Nuestro resultado {silhouette_avg:.3f} indica: ", end="")
if silhouette_avg > 0.7:
    print("¡EXCELENTE separación!")
elif silhouette_avg > 0.5:
    print("Buena separación")
elif silhouette_avg > 0.3:
    print("Separación regular")
else:
    print("Separación pobre")

print("\n" + "="*60)

# 4. VISUALIZACIÓN PRINCIPAL
print("📈 PASO 4: Visualizando resultados...")
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Gráfico de clusters verdaderos
scatter1 = ax1.scatter(X[:, 0], X[:, 1], c=y_true, cmap='viridis', alpha=0.7)
ax1.set_title('Clusters Verdaderos', fontsize=14, fontweight='bold')
ax1.set_xlabel('Característica 1')
ax1.set_ylabel('Característica 2')
ax1.grid(True, alpha=0.3)
plt.colorbar(scatter1, ax=ax1)

# Gráfico de resultados K-means
scatter2 = ax2.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', alpha=0.7)
ax2.scatter(cluster_centers[:, 0], cluster_centers[:, 1], 
           c='red', marker='x', s=300, linewidths=4, label='Centroides')
ax2.set_title('Resultados K-means', fontsize=14, fontweight='bold')
ax2.set_xlabel('Característica 1')
ax2.set_ylabel('Característica 2')
ax2.grid(True, alpha=0.3)
ax2.legend()
plt.colorbar(scatter2, ax=ax2)

plt.tight_layout()
plt.show()

# 5. MÉTODO DEL CODO (ELBOW METHOD)
print("📊 PASO 5: Aplicando el Método del Codo para encontrar k óptimo...")
k_range = range(1, 11)
inertias = []
silhouette_scores = []

for k in k_range:
    if k == 1:
        # Silhouette no se puede calcular con 1 cluster
        inertias.append(KMeans(n_clusters=k, random_state=42, n_init=10).fit(X).inertia_)
        silhouette_scores.append(0)
    else:
        kmeans_temp = KMeans(n_clusters=k, random_state=42, n_init=10)
        y_pred_temp = kmeans_temp.fit_predict(X)
        inertias.append(kmeans_temp.inertia_)
        silhouette_scores.append(silhouette_score(X, y_pred_temp))

# Gráficos del método del codo y silueta
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Método del Codo
ax1.plot(k_range, inertias, 'bo-', linewidth=2, markersize=8)
ax1.set_xlabel('Número de Clusters (k)', fontsize=12)
ax1.set_ylabel('Inercia', fontsize=12)
ax1.set_title('Método del Codo', fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.axvline(x=4, color='red', linestyle='--', alpha=0.7, label='k=4 (óptimo)')
ax1.legend()

# Método de la Silueta
ax2.plot(k_range[1:], silhouette_scores[1:], 'ro-', linewidth=2, markersize=8)
ax2.set_xlabel('Número de Clusters (k)', fontsize=12)
ax2.set_ylabel('Puntuación Promedio de Silueta', fontsize=12)
ax2.set_title('Método de la Silueta', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3)
# Encontrar el k con mejor silueta
best_k = k_range[1:][np.argmax(silhouette_scores[1:])]
ax2.axvline(x=best_k, color='red', linestyle='--', alpha=0.7, label=f'k={best_k} (máxima silueta)')
ax2.legend()

plt.tight_layout()
plt.show()

print("💡 INTERPRETACIÓN DE LOS MÉTODOS:")
print("🔹 MÉTODO DEL CODO:")
print("   - Busca el punto donde la inercia deja de decrecer significativamente")
print("   - El 'codo' en la curva indica el k óptimo")
print(f"   - En este caso, k=4 muestra el codo más claro")

print("\n🔹 MÉTODO DE LA SILUETA:")
print("   - Busca el k que maximiza la puntuación promedio de silueta")
print(f"   - El k óptimo según silueta es: k={best_k}")
print(f"   - Puntuación máxima: {max(silhouette_scores[1:]):.3f}")

# 6. ANÁLISIS DETALLADO DE SILUETA - EXPLICACIÓN PARA DUMMIES
print(f"\n📊 PASO 6: ¿Qué es la Silueta? - Explicación súper simple")

print("🤔 IMAGINE QUE USTED ES UN PUNTO EN EL ESPACIO:")
print("   1. Mire a sus 'vecinos' en SU cluster → calcule distancia promedio = 'a'")
print("   2. Mire al cluster MÁS CERCANO → calcule distancia promedio = 'b'")
print("   3. Su valor de silueta = (b - a) / max(a, b)")
print("")
print("🎯 ¿QUÉ SIGNIFICA ESTO?")
print("   • Si b >> a → Silueta cerca de +1 → ¡ESTÁ MUY BIEN UBICADO!")
print("   • Si b ≈ a  → Silueta cerca de 0  → Está en la 'frontera'")
print("   • Si b << a → Silueta cerca de -1 → ¡ESTÁ MAL UBICADO!")
print("")

# Demostración con un punto específico
punto_ejemplo = X[0]
cluster_ejemplo = y_pred[0]
print(f"🔍 EJEMPLO PRÁCTICO con el punto {punto_ejemplo.round(2)}:")
print(f"   - Está asignado al cluster {cluster_ejemplo}")

# Calcular 'a' - distancia promedio dentro de su cluster
puntos_mismo_cluster = X[y_pred == cluster_ejemplo]
distancias_mismo_cluster = np.sqrt(np.sum((puntos_mismo_cluster - punto_ejemplo)**2, axis=1))
a = np.mean(distancias_mismo_cluster)

# Calcular 'b' - distancia promedio al cluster más cercano
distancias_otros_clusters = []
for cluster_id in range(k):
    if cluster_id != cluster_ejemplo:
        puntos_otro_cluster = X[y_pred == cluster_id]
        dist_promedio = np.mean(np.sqrt(np.sum((puntos_otro_cluster - punto_ejemplo)**2, axis=1)))
        distancias_otros_clusters.append(dist_promedio)

b = min(distancias_otros_clusters)
silueta_ejemplo = (b - a) / max(a, b)

print(f"   - Distancia promedio a SU cluster (a): {a:.2f}")
print(f"   - Distancia promedio al cluster MÁS CERCANO (b): {b:.2f}")
print(f"   - Su valor de silueta: ({b:.2f} - {a:.2f}) / {max(a,b):.2f} = {silueta_ejemplo:.3f}")

if silueta_ejemplo > 0.5:
    print("   - ✅ ¡Este punto está MUY bien ubicado!")
elif silueta_ejemplo > 0:
    print("   - ✅ Este punto está bien ubicado")
else:
    print("   - ❌ Este punto podría estar mejor en otro cluster")

print("\n" + "="*70)
print("📊 AHORA VEAMOS TODOS LOS PUNTOS...")

sample_silhouette_values = silhouette_samples(X, y_pred)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))

# Gráfico de silueta CON EXPLICACIONES DETALLADAS
y_lower = 10
colors = plt.cm.viridis(np.linspace(0, 1, k))

print(f"📈 LEYENDO EL GRÁFICO DE SILUETA:")
for i in range(k):
    cluster_silhouette_values = sample_silhouette_values[y_pred == i]
    
    size_cluster_i = cluster_silhouette_values.shape[0]
    
    # Verificar que el cluster no esté vacío
    if size_cluster_i == 0:
        print(f"🔹 Cluster {i}: ¡VACÍO! (0 puntos asignados)")
        print(f"   - ⚠️  Este cluster no tiene puntos asignados")
        continue
        
    cluster_silhouette_values.sort()
    y_upper = y_lower + size_cluster_i
    
    # Estadísticas del cluster (solo si no está vacío)
    min_sil = cluster_silhouette_values.min()
    max_sil = cluster_silhouette_values.max()
    avg_sil = cluster_silhouette_values.mean()
    
    print(f"🔹 Cluster {i}: {size_cluster_i} puntos")
    print(f"   - Silueta mín/máx/promedio: {min_sil:.3f} / {max_sil:.3f} / {avg_sil:.3f}")
    
    # Evaluar calidad del cluster
    puntos_negativos = np.sum(cluster_silhouette_values < 0)
    if puntos_negativos > 0:
        print(f"   - ⚠️  {puntos_negativos} puntos con silueta negativa (mal ubicados)")
    else:
        print(f"   - ✅ Todos los puntos bien ubicados")
    
    if avg_sil > silhouette_avg:
        print(f"   - 👍 Cluster MEJOR que el promedio")
    else:
        print(f"   - 👎 Cluster PEOR que el promedio")
    
    ax1.fill_betweenx(np.arange(y_lower, y_upper),
                      0, cluster_silhouette_values,
                      facecolor=colors[i], edgecolor=colors[i], alpha=0.7)
    
    # Etiqueta del cluster con estadísticas
    ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, 
             f'C{i}\n({size_cluster_i}p)', fontsize=10, fontweight='bold')
    y_lower = y_upper + 10

ax1.set_xlabel('Valores de Silueta\n(Más a la derecha = Mejor ubicado)', fontsize=12)
ax1.set_ylabel('Puntos agrupados por Cluster', fontsize=12)
ax1.set_title('🎯 Gráfico de Silueta - ¿Qué tan bien están ubicados los puntos?', 
              fontsize=13, fontweight='bold')

# Línea vertical para el promedio CON EXPLICACIÓN
ax1.axvline(x=silhouette_avg, color="red", linestyle="--", linewidth=3,
            label=f'Promedio = {silhouette_avg:.3f}')
ax1.axvline(x=0, color="orange", linestyle=":", linewidth=2,
            label='Límite Bueno/Malo')

# Agregar zonas de interpretación
ax1.axvspan(-1, 0, alpha=0.1, color='red', label='Zona MALA')
ax1.axvspan(0, 0.3, alpha=0.1, color='yellow', label='Zona REGULAR')
ax1.axvspan(0.3, 1, alpha=0.1, color='green', label='Zona BUENA')

ax1.legend(loc='lower right')
ax1.grid(True, alpha=0.3)
ax1.set_xlim(-0.3, 1)

# Gráfico de clusters con colores de silueta Y EXPLICACIONES
scatter = ax2.scatter(X[:, 0], X[:, 1], c=sample_silhouette_values, 
                     cmap='RdYlGn', alpha=0.8, s=60)  # RdYlGn: rojo=malo, amarillo=regular, verde=bueno
ax2.scatter(cluster_centers[:, 0], cluster_centers[:, 1], 
           c='blue', marker='x', s=400, linewidths=5, label='Centroides')

# Marcar puntos problemáticos
puntos_malos = sample_silhouette_values < 0
if np.any(puntos_malos):
    ax2.scatter(X[puntos_malos, 0], X[puntos_malos, 1], 
               s=100, facecolors='none', edgecolors='red', linewidths=2,
               label=f'{np.sum(puntos_malos)} puntos mal ubicados')

ax2.set_title('🎨 Mapa de Calidad: Verde=Bien, Amarillo=Regular, Rojo=Mal', 
              fontsize=13, fontweight='bold')
ax2.set_xlabel('Característica 1')
ax2.set_ylabel('Característica 2')
ax2.legend()

# Barra de colores con interpretación
cbar = plt.colorbar(scatter, ax=ax2, label='Valor de Silueta')
cbar.ax.set_ylabel('🔴 Mal ubicado ← → Bien ubicado 🟢', rotation=270, labelpad=20)

plt.tight_layout()
plt.show()

print("\n💡 RESUMEN PARA DUMMIES - ¿CÓMO LEER ESTOS GRÁFICOS?")
print("="*70)
print("📊 GRÁFICO DE LA IZQUIERDA (Gráfico de Silueta):")
print("   🔹 Cada 'montañita' horizontal = un cluster")
print("   🔹 Altura de la montañita = número de puntos en ese cluster")
print("   🔹 Ancho hacia la derecha = qué tan bien ubicados están los puntos")
print("   🔹 Si una montañita cruza la línea roja = cluster mejor que promedio")
print("   🔹 Si hay partes a la IZQUIERDA del 0 = puntos mal ubicados")
print("")
print("🎨 GRÁFICO DE LA DERECHA (Mapa de Colores):")
print("   🟢 Verde = Puntos felices en su cluster")
print("   🟡 Amarillo = Puntos 'meh', podrían estar mejor")
print("   🔴 Rojo = Puntos tristes, están en el cluster equivocado")
print("   💙 X Azules = Centros de cada cluster")
print("   🔴 Círculos rojos = Puntos definitivamente mal ubicados")

# Dar recomendaciones basadas en los resultados
puntos_negativos_total = np.sum(sample_silhouette_values < 0)
print(f"\n🎯 DIAGNÓSTICO DE NUESTRO CLUSTERING:")
if puntos_negativos_total == 0:
    print("✅ ¡PERFECTO! Ningún punto mal ubicado")
elif puntos_negativos_total < len(X) * 0.05:  # Menos del 5%
    print(f"✅ MUY BUENO: Solo {puntos_negativos_total} puntos mal ubicados ({puntos_negativos_total/len(X)*100:.1f}%)")
elif puntos_negativos_total < len(X) * 0.1:   # Menos del 10%
    print(f"⚠️  BUENO: {puntos_negativos_total} puntos mal ubicados ({puntos_negativos_total/len(X)*100:.1f}%)")
else:
    print(f"❌ REGULAR: Muchos puntos mal ubicados ({puntos_negativos_total/len(X)*100:.1f}%)")
    print("   💡 Sugerencia: Probar con diferente número de clusters")

if silhouette_avg > 0.7:
    print("🏆 Calidad general: EXCELENTE")
elif silhouette_avg > 0.5:
    print("👍 Calidad general: BUENA") 
elif silhouette_avg > 0.3:
    print("🤷 Calidad general: REGULAR")
else:
    print("👎 Calidad general: POBRE - considere cambiar k")

print("="*70)

# 7. PREDICCIÓN EN NUEVOS PUNTOS
print(f"\n🎯 PASO 7: Prediciendo clusters para nuevos puntos...")
new_points = np.array([[2, 2], [-8, 3], [6, -6]])
predicted_clusters = kmeans.predict(new_points)

print(f"📍 Nuevos puntos: {new_points}")
print(f"🎯 Clusters predichos: {predicted_clusters}")

# Calcular distancias a centroides para explicar la asignación
print("\n💡 EXPLICACIÓN DE LAS PREDICCIONES:")
for i, point in enumerate(new_points):
    distances = np.sqrt(np.sum((cluster_centers - point)**2, axis=1))
    closest_centroid = np.argmin(distances)
    print(f"📍 Punto {i+1} {point}:")
    print(f"   - Asignado al cluster {predicted_clusters[i]}")
    print(f"   - Distancia al centroide más cercano: {distances[closest_centroid]:.2f}")

# Visualización final con nuevos puntos
plt.figure(figsize=(10, 8))
scatter = plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', alpha=0.6, s=50)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], 
           c='red', marker='x', s=300, linewidths=4, label='Centroides')
plt.scatter(new_points[:, 0], new_points[:, 1], 
           c='black', marker='s', s=200, label='Nuevos Puntos', edgecolors='white', linewidths=2)

# Anotar los nuevos puntos
for i, point in enumerate(new_points):
    plt.annotate(f'P{i+1}\n(Cluster {predicted_clusters[i]})', 
                xy=point, xytext=(10, 10), textcoords='offset points',
                bbox=dict(boxstyle='round,pad=0.3', fc='yellow', alpha=0.7),
                fontsize=10, fontweight='bold')

plt.title('K-means: Clusters y Predicciones en Nuevos Puntos', fontsize=14, fontweight='bold')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.legend()
plt.colorbar(scatter, label='Cluster')
plt.grid(True, alpha=0.3)
plt.show()

print("\n" + "="*60)
print("🎉 RESUMEN DEL ANÁLISIS:")
print(f"✅ Número óptimo de clusters: {k}")
print(f"✅ Calidad del clustering (ARI): {ari_score:.3f}")
print(f"✅ Separación de clusters (Silueta): {silhouette_avg:.3f}")
print(f"✅ Compacidad de clusters (Inercia): {inertia:.2f}")
print("✅ El algoritmo funcionó correctamente identificando los 4 grupos naturales")
print("="*60)
```
